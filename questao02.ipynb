Questão 2: [1/2 ponto]

Cinco programadores estão trabalhando em um laboratório. Cada um está desenvolvendo um módulo
diferente de um grande sistema, mas todos precisam compilar seus códigos com recursos limitados:
* Um compilador
* um banco de dados de dependências compartilhado.


---


Cada programador:

* Precisa adquirir acesso exclusivo ao compilador.
* Precisa de acesso compartilhado ao banco de dados, mas apenas dois programadores podem
acessá-lo simultaneamente para evitar sobrecarga.

* Apôs compilar, o programador descansa (pensa) por algum tempo e depois tenta compilar
novamente.
---


Regras:
* Apenas um programador pode usar o compilador por vez.
* No máximo dois programadores podem acessar o banco de dados ao mesmo tempo.
* Um programador só pode começar a compilação quando tiver ambos os recursos.
* O sistema deve evitar deadlocks e inanição.


---


Apresente a saída de forma que seja possível identificar a atividade ou estado de momento de cada
programador.

Faça com que execute em laço infinito para sua apresentação em sala.


```python
import threading
import time
import random

PROGRAMADORES = 5 #Quantidade de programadores

stop_flag = True #Flag de parada, para matar as threads ao encerrar o programa
```

```python
def programador(id):
  while not stop_flag: #Loop até a bandeira de parada ser acionada
    time.sleep(random.uniform(1, 4)) #Espera antes de rodar
    # --- Área Crítica do banco ---
    print(f"\n{id} quer usar o banco, esperando...")
    bank.acquire() #Tenta entrar, se não houver vaga, espera até liberar
    print(f"\n{id} está usando o banco")

    # --- Área Crítica do compilador ---
    print(f"\n{id} quer usar o compilador, esperando...")
    with compilator: #Mutex do python, que ao entrar da .acquire e ao sair .release na região critica relacionada
      print(f"\n{id} entrou no compilador")
      print(f"{id} está compilando")
      time.sleep(random.uniform(2, 5)) #Tempo de compilação
      print(f"\n{id} terminou de usar o compilador")
      print("compilador liberado")

    print(f"\n{id} terminou de usar o banco")
    print("banco liberado")
    bank.release() #Libera a vaga

    print(f"\n{id} está pensando")
    time.sleep(random.uniform(1, 4)) #Pensando antes de executar de novo
  print(f"\n{id} finalizado")
```

```python
#Cria os semaforosantes de executar, para ter certeza que estão vazios
compilator = threading.Lock()
bank = threading.Semaphore(2)

stop_flag = False #Libera a flag para os programadores poderem rodar

threads = [] #Lista das threads
for i in range(PROGRAMADORES): #Criação e inicialização das threads
    t = threading.Thread(target=programador, args=(f"Programdor-{i+1}",), daemon=True)
    threads.append(t)
    t.start()

try:
    while True:
        time.sleep(1) #Mantém a thread principal viva

except KeyboardInterrupt:
    stop_flag = True #Trava a flag para os programadores pararem de rodar
    for t in threads: #Espera todas as threads terminarem
        t.join()
    print("\nSimulação encerrada.")
```
